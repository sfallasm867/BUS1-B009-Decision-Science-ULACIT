```python
import pandas as pd

df = pd.DataFrame()

# creamos las columnas
df['Decision'] = None
df['Good Conditions'] = None
df['Poor Conditions'] = None

# añadimos filas por su nombre de fila
df.loc[1] = ['Apartment Building', 50000, 30000]
df.loc[2] = ['Office Building', 100000, -40000]
df.loc[3] = ['Warehouse', 30000, 10000]

print(df)
```

```python
def maximo(valores):
    mayor = valores[0]

    for i in range(1, len(valores)):
        if valores[i] > mayor:
            mayor = valores[i]

    return mayor

numeros = [50000, 100000, 30000, 30000, -40000, 10000]

print(maximo(numeros))
```

```python
def minimo(valores):
    menor = valores[0]

    for i in range(1, len(valores)):
        if valores[i] < menor:
            menor = valores[i]

    return menor

numeros = [50000, 100000, 30000, 30000, -40000, 10000]

print(minimo(numeros))
```
